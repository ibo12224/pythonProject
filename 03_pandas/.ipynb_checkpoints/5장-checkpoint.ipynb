```python
weather=pd.read_csv('data/weather.csv')
weather.head(1)
```

```python
weather_melt=weather.melt(['id','year','month','element'])
```

```python
weather_melt.columns=['id', 'year', 'month', 'element', 'day', 'temp']
```

```python
weather_tidy=weather_melt.pivot_table(index=['id', 'year', 'month', 'day'],
                        columns=['element'],values=['temp'])
```

```python
weather_tidy.reset_index(inplace=True)
weather_tidy
```

```python
result=(weather_melt.pivot_table(index=['id', 'year', 'month', 'day'],
                        columns=['element'],values=['temp'])
        .reset_index())
result
```
